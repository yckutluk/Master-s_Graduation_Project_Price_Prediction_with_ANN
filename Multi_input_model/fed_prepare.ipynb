```python
#importing the fed data
df=pd.read_csv('fed.csv')
df
```

```python

sütun_degisimleri = {
    'Tarih': 'Date',
    'Şimdi': 'Currently',
    'Açılış': 'Open',
    'Yüksek': 'High',
    'Düşük': 'Low',
    'Hac.': 'Vol',
    'Fark %': 'Dif',
    
    
    # ve daha fazla sütun değişimi
}
```

```python
df.rename(columns=sütun_degisimleri, inplace=True)
```

```python
df
```

```python
df = df[['Date', 'Currently']]
df
```

```python
# 'Date' sütununu datetime veri tipine dönüştürme
df['Date'] = pd.to_datetime(df['Date'], format='%d.%m.%Y')

# 'Date' sütununu indeks olarak ayarlama
df.set_index('Date', inplace=True)

# 'yıl-ay-gün' formatına dönüştürme
df.index = df.index.strftime('%Y-%m-%d')



```

```python
# Virgülü nokta ile değiştirme
df['Currently'] = df['Currently'].str.replace(',', '.')

# Sütunları float64 veri tipine dönüştürme
df['Currently'] = df['Currently'].astype(float)
```
